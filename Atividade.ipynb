### Atividade Complementar


Imagine que você é o responsável pelo controle de estoque e pretende criar um sistema que registre os pedidos das pessoas ou informe que não há estoque suficiente para atender a demanda.

Nesse caso digamos que os produtos que estão disponíveis no estoque sejam Sabão (com 75 unidades disponíveis); Detergente (com 30 unidades disponíveis) e Desinfetante (com 50 unidades disponíveis).

Crie um código em que seja pedido ao usuário que informe o produto e a quantidade desejada.

Ao final o sistema precisa analisar o pedido com os valores disponíveis no estoque e informar se o pedido foi feito com sucesso ou não foi possível atender o pedido devido à quantidade em estoque ser inferior à demandada.

```python
sabao = 75
detergente = 30
desinfetante = 50

produto = input("Qual produto deseja? (Sabão, Detergente ou Desinfetante): ")
quantidade = int(input("Quantidade desejada: "))
if produto != "Sabão" or "Detergente" or "Desinfetante":
  if produto == "Sabão":
    if quantidade <= sabao:
      print("Pedido realizado com sucesso!")
    else:
      print("Não foi possível atender o pedido: estoque insuficiente.")
  if produto == "Detergente":
      if quantidade <= detergente:
        print("Pedido realizado com sucesso!")
      else:
        print("Não foi possível atender o pedido: estoque insuficiente.")
  if produto == "Desinfetante":
    if quantidade <= desinfetante:
      print("Pedido realizado com sucesso!")
else:
  print("Não foi possível atender o pedido: estoque insuficiente.")
```

7 - Crie linhas de código em que o usuário deve informar apenas os números do CPF (que contém 11 caracteres). Neste caso o usuário poderá informar o CPF no formato que quiser, contanto que ao final haja uma codificação de CPF válida (11 caracteres numéricos).

Resumindo: O usuário pode informar o CPF no formato que desejar, mas, no momento da exibição na tela o CPF informado deve estar no formato xxx.xxx.xxx-xx. Caso o que o usuário informar não resulte no formato padrão dos CPF (xxx.xxx.xxx-xx) uma mensagem de erro deverá ser informada na tela.

```python
entrada = input("Informe o CPF (em qualquer formato): ")
numeros = ""
for caractere in entrada:
  if caractere.isdigit():
    numeros = numeros + caractere
if len(numeros) == 11:
  cpf_formatado = f"{numeros[0:3]}.{numeros[3:6]}.{numeros[6:9]}-{numeros[9:11]}"
  print(f"CPF: {cpf_formatado}")
else:
  print("Erro: CPF inválido. Devem existir 11 números.")
```

Nessas linhas de código o usuário poderá optar em informar ou não o primeiro dígito obrigatório ("9"), mas, ao final, seu programa deve sempre informar na tela o número do celular no formato padrão 9xxxx-xxxx.

```python
entrada = input("Informe o número do celular (somente números): ")
numeros = ""
for c in entrada:
  if c.isdigit():
    numeros = numeros + c
if len(numeros) == 8: # usuário omitiu o "9" obrigatório
  numeros = "9" + numeros
if len(numeros) == 9 and numeros[0] == "9":
  celular = f"{numeros[0:5]}-{numeros[5:9]}" # 9xxxx-xxxx
  print(f"Celular: {celular}")
else:
  print("Erro no preenchimento! O celular deve ter 9 dígitos e começar com 9.")
```

2 - Crie linhas de código que permitam que o usuário informe o valor do seu salário e de suas despesas no padrão monetário brasileiro (por exemplo Salário = 5.520,45 e Despesas = 3.895,86). A partir dessas informações seu sistema deve apurar o "Resultado Financeiro" (que é o resultado da diferença entre o valor do "Salário" e o valor das "Despesas") e também apurar a "Situação Financeira" (que será considerada "Boa", quando o Resultado Financeiro for maior ou igual a zero, e "Ruim", quando o Resultado Financeiro for negativo).

Ao final o seu sistema deve mostrar na tela as seguintes informações:

Salário: R$\$$ 5.520,45

Despesas: R$ 3.895,86

Resultado Financeiro: R$\$$ 1.624,59

Situação Financeira: Boa

```python
salario_str = input("Informe o salário (ex: 5.520,45): ")
despesas_str = input("Informe as despesas (ex: 3.895,86): ")
salario = float(salario_str.replace(".", "").replace(",", "."))
despesas = float(despesas_str.replace(".", "").replace(",", "."))
resultado = salario - despesas
if resultado >= 0:
  situacao = "Boa"
else:
  situacao = "Ruim"
def br(n):
  return f"{n:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
print(f"Salário: R$ {br(salario)}")
print(f"Despesas: R$ {br(despesas)}")
print(f"Resultado Financeiro: R$ {br(resultado)}")
print(f"Situação Financeira: {situacao}")
```

3 - Aproveitando a questão anterior (Questão 2), crie linhas de código que faça extamente o que se pede na Questão 2, porém, desta vez o seu sistema deve possuir um código de controle de validação que emita uma mensagem de erro quando o usuário não preencher (deixar em branco) o Salário ou as Despesas ou caso o usuário preencha valores que não correspondam ao padrão monetário brasileiro (por exemplo colocar letras ou caracteres especiais no lugar dos valores numéricos).

Assim, apenas uma das duas situação abaixo deve aparecer na tela para ao usuário:

QUANDO O USUÁRIO NÃO CUMPRIR AS REGRAS DE VALIDAÇÃO:
Erro no preenchimento! Favor informar valores numéricos válidos!

QUANDO O USUÁRIO CUMPRIR AS REGRAS DE VALIDAÇÃO:
Salário: R$\$$ 5.520,45
Despesas: R$\$$ 3.895,86
Resultado Financeiro: R$\$$ 1.624,59
Situação Financeira: Boa

```python
salario_str = input("Informe o salário (ex: 5.520,45): ")
despesas_str = input("Informe as despesas (ex: 3.895,86): ")

def converter_br(texto):
    return float(texto.replace(".", "").replace(",", "."))

def br(n):
    return f"{n:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

try:
    salario = converter_br(salario_str)
    despesas = converter_br(despesas_str)
    resultado = salario - despesas
    if resultado >= 0:
        situacao = "Boa"
    else:
        situacao = "Ruim"
    print(f"Salário: R$ {br(salario)}")
    print(f"Despesas: R$ {br(despesas)}")
    print(f"Resultado Financeiro: R$ {br(resultado)}")
    print(f"Situação Financeira: {situacao}")
except ValueError:
    print("Erro no preenchimento! Favor informar valores numéricos válidos!")
```

### Digamos que o setor onde você trabalha quer instituir um controle de almoxarifado e possui as seguintes informações: produtos = ["canettta", "mouseeeeee", "lapis", "borracha", "grampeador", "cartuxxo"] quantidades = [250, 50, 320, 450, 70, 190] Com base nessas informações responda as questões 5 a 10:

5 - Crie um código que faça a correção dos erros de digitação ocorridos na lista "produtos".

```python
produtos = ["canettta", "mouseeeeee", "lapis", "borracha", "grampeador", "cartuxxo"]
quantidades = [250, 50, 320, 450, 70, 190]
produtos[0] = "caneta"
produtos[1] = "mouse"
produtos[5] = "cartucho"
print(produtos)
```

6 - Crie um sistema onde seja possível um usuário consultar quantas unidades de um produto constam no almoxarifado. Nesse sistema o usuário informa o nome do produto e o sistema deve informar o seguinte:

Há XX unidades do produto YY no almoxarifado.

importante: considere que o usuário sempre informará um produto que consta no estoque.

```python
consulta = input("Qual produto deseja consultar? ")
posicao = produtos.index(consulta)
print(f"Há {quantidades[posicao]} unidades do produto {consulta} no almoxarifado.")
```

7 - Semelhante à questão 6, crie um sistema onde seja possível um usuário informar o produto que ele deseja consultar no almoxarifado. Nesse sistema o usuário informa o nome do produto e o sistema deve se comportar das seguintes formas:

Quando o produto informado existir no almoxarifado, informar: "Há XX unidades do produto YY no almoxarifado."

Quando o produto informado não existir no almoxarifado ou o usuário informar um produto com erro de digitação: "O produto ZZ não consta no almoxarifado."


```python
consulta = input("Qual produto deseja consultar? ")
if consulta in produtos:
  posicao = produtos.index(consulta)
  print(f"Há {quantidades[posicao]} unidades do produto {consulta} no almoxarifado.")
else:
  print(f"O produto {consulta} não consta no almoxarifado.")
```

8 - Crie linhas de código que permitam identificar qual o produto que possui mais unidades e qual possui menos unidades no almoxarifado.

```python
maior = max(quantidades)
menor = min(quantidades)
produto_maior = produtos[quantidades.index(maior)]
produto_menor = produtos[quantidades.index(menor)]
print(f"Produto com MAIS unidades: {produto_maior} ({maior} unidades)")
print(f"Produto com MENOS unidades: {produto_menor} ({menor} unidades)")
```

9 - Crie um código que permita saber a quantidade total de produtos que há no almoxarifado, ou seja, o sistema deve somar todos os itens da lista "quantidades".

10 - Digamos que o setor não utilizará mais o produto "cartucho" e irá trocar com outro setor todos esses "cartuchos" por 30 unidades de "teclado" para computadores. ### EXERCÍCIO COMPLEMENTAR

Digamos que você trabalha num setor que atende o público por meio de processos administrativos. As listas representam os meses e as quantidades de processos atendidos, respectivamente:

meses = ["janeiro", "fevereiro", "março", "abril", "maio", "junho", "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"]

atendimentos_semestre_1 = [350, 280, 420, 560, 325, 295]
atendimentos_semestre_2 = [490, 550, 620, 516, 380, 478]

Crie um sistema onde seja possível evidenciar na tela o meses em que ocorreram as maiores e menores quantidades de atendimentos, bem como evidenciar o percentual que esses atendimentos representam em relação ao total de atendimentos no ano.

No sistema deve ser evidenciada as seguintes menagens:


"O mês de xxx foi o que registrou a maior quantidade de atendimentos no ano com yyy atendimentos, o que representa zz % do total de atendimentos no ano."

"O mês de xxx foi o que registrou a menor quantidade de atendimentos no ano com yyy atendimentos, o que representa apenas zz % do total de atendimentos no ano."

```python
meses = ["janeiro", "fevereiro", "março", "abril", "maio", "junho",
"julho", "agosto", "setembro", "outubro", "novembro", "dezembro"]
atendimentos_semestre_1 = [350, 280, 420, 560, 325, 295]
atendimentos_semestre_2 = [490, 550, 620, 516, 380, 478]
atendimentos = atendimentos_semestre_1 + atendimentos_semestre_2
total_ano = sum(atendimentos)
maior = max(atendimentos)
mes_maior = meses[atendimentos.index(maior)]
perc_maior = maior / total_ano * 100
menor = min(atendimentos)
mes_menor = meses[atendimentos.index(menor)]
perc_menor = menor / total_ano * 100
print(f"O mês de {mes_maior} foi o que registrou a maior quantidade de "
f"atendimentos no ano com {maior} atendimentos, o que representa "
f"{perc_maior:.2f}% do total de atendimentos no ano.")
print(f"O mês de {mes_menor} foi o que registrou a menor quantidade de "
f"atendimentos no ano com {menor} atendimentos, o que representa "
f"apenas {perc_menor:.2f}% do total de atendimentos no ano.")
```
